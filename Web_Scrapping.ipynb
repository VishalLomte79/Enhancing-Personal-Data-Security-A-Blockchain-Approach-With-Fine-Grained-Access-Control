```python


import requests
from bs4 import BeautifulSoup

# Step 3: Define the URL of the website you want to scrape
url = 'https://techcrunch.com/2024/04/30/sams-clubs-ai-powered-exit-tech-reaches-20-of-stores/'  # Replace this with the actual website URL

# Step 4: Send a request to fetch the HTML content of the webpage
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Step 5: Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Step 6: Extract and print the desired content
    # For this example, let's print all the text in <p> tags
    paragraphs = soup.find_all('p')

    # Print the text content of each <p> tag
    for paragraph in paragraphs:
        print(paragraph.get_text())
else:
    print(f"Failed to retrieve content. Status code: {response.status_code}")
```
